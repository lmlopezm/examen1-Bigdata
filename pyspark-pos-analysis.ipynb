```python
import nltk
from pyspark import SparkContext
import string
import re
```

```python
nltk.download("punkt")
nltk.download("averaged_perceptron_tagger")
```

```python
sc = SparkContext("spark://lau-VirtualBox:7077", appName = "pyspark-pos-analysis")
```

```python
# Loading a text file
rdd_reuters = sc.textFile("./data/reuters.txt")
```

```python
rdd_reuters.count()
```

```python
type(u'lama')
string.punctuation
punc = '!"#$%&\'()*+,./:;<=>?@[\\]^_`{|}~"'

def limpiar_texto(x):
    converted = x
    lowercased_str = converted.lower()
    #lowercased_str = lowercased_str.replace(",","")
    lowercased_str = lowercased_str.replace("-"," ")
    lowercased_str = re.sub(r"[^a-zA-Z-' ]", "", lowercased_str)
    #lowercased_str = lowercased_str.replace("!","")
    #lowercased_str = lowercased_str.replace(".","")
    #lowercased_str = lowercased_str.replace("...","")
    #lowercased_str = lowercased_str.replace(";","")
    # for more difficult cases use re.split('A|B')
    lowercased_str = lowercased_str.replace('--',' ')
    clean_str = lowercased_str.translate(punc) #change 1
    
    return clean_str
    
```

```python
def listToString(s): 
    
    # initialize an empty string
    str1 = " " 
    
    # traverse in the string  
    for ele in s: 
        str1 += ele + " "  
    
    # return string  
    return str1 

    
```

```python
#one_RDD.collect()

one_RDD = rdd_reuters.flatMap(lambda x: limpiar_texto(x).split())
one_RDD = one_RDD.map(lambda x: (x,1))
one_RDD = one_RDD.reduceByKey(lambda x,y: x + y)
s = [i for i,j in one_RDD.collect()]

```

```python
newlist = []
for i in s:
    newlist.append(i)
newlist
```

```python

listToString(s)
```

```python
text = nltk.word_tokenize(listToString(s))
nltk.pos_tag(text)
```
